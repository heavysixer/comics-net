## MetaData

```python
import glob
import pandas as pd

path =r'../metadata' # use your path
all_files = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in all_files:
    df = pd.read_csv(file_, index_col=0, header=0)
    list_.append(df)
df = pd.concat(list_).reset_index(drop=True)
```

```python
df
```

```python
from ast import literal_eval

metadata = df['metadata'].apply(lambda x: literal_eval(x)).reset_index()

md_idx = metadata.loc[metadata['metadata'] != {}, 'index']
md = metadata.loc[metadata['metadata'] != {}, 'metadata']
md = pd.DataFrame(md.tolist(), index=md_idx)
md = md.loc[md['characters'].notnull(), ['characters', 'pencils', 'colors', 'inks', 'letters']].reset_index()
md.head()
```

## Dedupe Issues

```python
covers = os.listdir('../covers/')
covers = pd.DataFrame('./covers/' + pd.Series(covers))
covers.columns = ['cover']
```

```python
issue_data = pd.merge(covers, df, how='left').drop_duplicates()[['cover', 'series', 'right', 'title']].reset_index()
issue_data.columns = ['index', 'image_path', 'series', 'right', 'title']
```

```python
df = pd.merge(issue_data, md, on='index', how='inner')
df.head()
```

## Clean Rights

```python
def update_rights(t):
    if len(t) == 2:
        return t
    else:
        return ['NaN', t[0]]

rights = df['right'].fillna('NaN').apply(lambda x: re.sub(r"[^a-zA-Z0-9]+", ' ', x).strip())
rights = rights.apply(lambda x: x.split(' ')).apply(lambda x: pd.DataFrame(update_rights(x)))
rights = pd.concat(list(rights.values), axis=1).T.reset_index(drop=True)
rights.columns = ['right_month', 'right_year']
rights.head()
```

```python
df = pd.concat([df, rights], axis=1).drop('right', axis=1)
```

```python
from datetime import datetime

def string_to_date(s):
    try:
        return datetime.strptime(s, "%B, %Y")
    except:
        return np.nan

dates = map(lambda x: string_to_date(x), 
    rights['right_month'] + ', ' + rights['right_year'])
min_date = pd.Series(dates).dropna().min()
months_old = map(lambda x: np.round((x - min_date).days / 30.42, 0), pd.Series(dates))
months_old = pd.Series(months_old, name='months_old')
months_old.head()
```

```python
df = pd.concat([df, months_old], axis=1)
df.head()
```

## Title

```python
def update_title(t):
    if len(t) == 2:
        return [t[0], t[1].split(' ')[0]]
    else:
        return [t[0], 'NaN']
    
title = df['title'].str.split(' #').apply(lambda x: pd.DataFrame(update_title(x)))
title = pd.concat(list(title.values), axis=1).T.reset_index(drop=True)
title.columns = ['issue_name', 'issue_number']
```

```python
df = pd.concat([df, title], axis=1)
```
